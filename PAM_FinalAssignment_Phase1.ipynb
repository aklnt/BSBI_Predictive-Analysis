```python
eth = pd.read_csv('ethereum.csv')
eth.head()
```

```python
rows, columns = eth.shape
print(f"Ethereum dataset has {rows} rows and {columns} columns.")
```

```python
btc["Date"] = pd.to_datetime(btc["Date"])
eth["date"] = pd.to_datetime(eth["date"])
```

```python
eth.rename(columns={"date": "Date", "price": "Volume"}, inplace=True)
eth.head()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 5))
plt.plot(btc["Date"], btc["Close"], label="Bitcoin Closing Price", color="blue")
plt.xlabel("Date")
plt.ylabel("Price (USD)")
plt.title("Bitcoin Closing Price Over Time")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(12, 5))
plt.plot(eth["Date"], eth["Close"], label="Ethereum Closing Price", color="green")
plt.xlabel("Date")
plt.ylabel("Price (USD)")
plt.title("Ethereum Closing Price Over Time")
plt.legend()
plt.show()
```

```python
start_date = btc["Date"].min()
end_date = btc["Date"].max()

btc_filtered = btc[(btc["Date"] >= start_date) & (btc["Date"] <= end_date)]
eth_filtered = eth[(eth["Date"] >= start_date) & (eth["Date"] <= end_date)]
```

```python
plt.figure(figsize=(12, 5))

plt.plot(btc_filtered["Date"], btc_filtered["Close"], label="Bitcoin Closing Price", color="blue")

plt.plot(eth_filtered["Date"], eth_filtered["Close"], label="Ethereum Closing Price", color="green")

plt.xlabel("Date")
plt.ylabel("Price (USD)")
plt.title(f"Bitcoin & Ethereum Closing Prices ({start_date} to {end_date})")

plt.legend()

plt.show()
```

```python
btc.drop("Currency", axis=1, inplace=True)
eth.drop("Currency", axis=1, inplace=True)
```

```python
import seaborn as sns

correlation_matrix = btc.corr()

plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f")
plt.title('Correlation Matrix for Bitcoin Dataset')
plt.show()
```

```python
correlation_matrix = eth.corr()

plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f")
plt.title('Correlation Matrix for Ethereum Dataset')
plt.show()
```

```python
btc['Target'] = btc['Close'].shift(-1)  # Next day's closing price
eth['Target'] = eth['Close'].shift(-1)
```

```python
# Price change (Close - Open)
btc['Price_Change'] = btc['Close'] - btc['Open']
eth['Price_Change'] = eth['Close'] - eth['Open']

# Price spread (High - Low)
btc['Price_Spread'] = btc['High'] - btc['Low']
eth['Price_Spread'] = eth['High'] - eth['Low']

# Moving average features
btc['MA7'] = btc['Close'].rolling(window=7).mean()  # 7-day moving average
btc['MA30'] = btc['Close'].rolling(window=30).mean()  # 30-day moving average
eth['MA7'] = eth['Close'].rolling(window=7).mean() # 7-day moving average
eth['MA30'] = eth['Close'].rolling(window=30).mean() # 30-day moving average
```

```python
# Check for missing values after feature engineering
print("\nMissing Values in Bitcoin Dataset After Feature Engineering:")
print(btc.isnull().sum())
```

```python
btc.fillna(method='bfill', inplace=True)
eth.fillna(method='bfill', inplace=True)
```

```python
print("\nFinal Processed Bitcoin Dataset:\n", btc.head())
print("\nFinal Processed Ethereum Dataset:\n", eth.head())
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

btc[['Open', 'High', 'Low', 'Close', 'Volume', 'Price_Change', 'Price_Spread', 'MA7', 'MA30']] = scaler.fit_transform(
    btc[['Open', 'High', 'Low', 'Close', 'Volume', 'Price_Change', 'Price_Spread', 'MA7', 'MA30']])

eth[['Open', 'High', 'Low', 'Close', 'Volume', 'Price_Change', 'Price_Spread']] = scaler.fit_transform(
    eth[['Open', 'High', 'Low', 'Close', 'Volume', 'Price_Change', 'Price_Spread']])
```

```python
print("\nFinal Processed Bitcoin Dataset:\n", btc.head())
print("\nFinal Processed Ethereum Dataset:\n", eth.head())
```
